```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
x_train[0].shape
#dimension of an array
```

```python
x_train[0] #inputs of a handwritten numerical digit classified in binary 0 & 1, where 0 as white and 1 as black where range of visibility is in [0,255]
```

```python
plt.matshow(x_train[0])
```

```python
y_train[:7]
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train_flattened=x_train.reshape(len(x_train),28*28) #here, we flattened the 2D array matrix to 1D
x_test_flattened=x_test.reshape(len(x_test),28*28) #both, training and testing
```

```python
x_train_flattened.shape
```

```python
x_test_flattened.shape
```

```python
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),
    keras.layers.Dense(10, activation='sigmoid')
])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history = model.fit(x_train_flattened,y_train,epochs=5)
model.summary()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(history.history['accuracy'], history.history['loss'], marker='o')
plt.title('Loss vs. Accuracy during Training')
plt.xlabel('Accuracy')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

```python
model.evaluate(x_test_flattened,y_test)
```

```python
y_predicted=model.predict(x_test_flattened)
y_predicted[1]
```

```python
np.argmax(y_predicted[1])
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
y_test[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,9))
sns.heatmap(cm,annot=True,fmt='d')
plt.title('heatmap of confusion matrix of mnist digit predictions')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

```python
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),
    keras.layers.Dense(100, activation='relu'), #hidden layer added
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history = model.fit(x_train_flattened,y_train,epochs=5)
model.summary()
```
